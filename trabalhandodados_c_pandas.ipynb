# Manipulando Dados

## Criando novas colunas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
df = pd.read_csv(url)
df.head()
```

```python
df['Pontos_Extras'] = df['Notas'] * 0.4
df
```

```python
df.fillna(0, inplace=True)
df
```

```python
df['Notas_finais'] = df['Notas'] + df['Pontos_Extras']
df
```

```python
df['Aprovado_final'] = df['Notas_finais'].apply(lambda x: True if x >= 6 else False)
df
```

```python
selecao_nao_aprovados_antes_aprovados_depois = (df['Aprovado'] == False) & (df['Aprovado_final'] == True)
df[selecao_nao_aprovados_antes_aprovados_depois]
```

```python
df.to_csv('alunos_final.csv', index=False)
```
